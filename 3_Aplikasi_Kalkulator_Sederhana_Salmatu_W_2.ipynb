```python
panjang = int(input("Masukkan panjang dari persegi panjang (cm) : "))
lebar = int(input("Masukkan lebar dari persegi panjang (cm) : "))

luas = panjang * lebar

print (f"\nLuas dari persegi panjang yaitu {luas} cm^2")
```

#Kalkulator Menghitung Volume Bangun Kerucut

```python
r = int(input("Masukkan jari-jari dari bangun kerucut (cm) : "))
t = int(input("Masukkan tinggi bangun kerucut (cm) : "))

volume = 1/3 * 22/7 * r**2 * t

print(f"\nVolume dari bangun kerucut yaitu {volume} cm^3")
```
